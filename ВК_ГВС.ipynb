- Для систем горячего водоснабжения здания с одним теплообменником в ИТП для нескольких зон по высоте общий циркуляционный расход следует определять, как сумму циркуляционных расходов каждой зоны.

В.2 Расчет тепловой изоляции оборудования и трубопроводов
В практических расчетах тепловой изоляции принимается ряд допущений, позволяющих использовать упрощенные расчетные формулы.


```python
import math
import pandas as pd 

class Pipe_segment:
    pipes = []

    def __init__(self, dn_m, material, length_m=1, isolation_width_m: float = 0.025, isolation_lambda: float = 0.05, tout=18):
        self.dn_m = dn_m
        self.material = material
        self.length_m = length_m
        self.isolation_width_m = isolation_width_m
        self.isolation_lambda = isolation_lambda
        self.isolation_d2 = dn_m + (2 * isolation_width_m)
        self.K = 1.2 if self.material == "steel" else 1.7
        self.Rout_L = 0.1
        self.tin=65
        self.tout=tout
        self.q_L = round(self.heatloss_pipe_unit())
        self.Qloss_W = round(self.q_L * self.length_m)
        self.pipes.append(self)

    
    def heatloss_pipe_unit(self) -> float:
        
        Ris = ((1/(2*3.1415*self.isolation_lambda)))*(math.log(self.isolation_d2/self.dn_m))
        return self.K * (self.tin - self.tout) / (Ris + self.Rout_L)
```
