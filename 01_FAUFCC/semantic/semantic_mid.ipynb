```python

# требование из СП 59.13330 

'''
5.2.1 На стоянке (парковке) транспортных средств личного пользования, 
расположенной на участке около здания организации сферы услуг или внутри этого здания, 
следует выделять 10 % машино-мест (но не менее одного места) для людей с инвалидностью, 

в том числе количество специализированных расширенных машино-мест для транспортных средств инвалидов, передвигающихся на кресле-коляске, определять расчетом, при числе мест: 
- до 100 включительно……. 5 %, но не менее одного места;
- от 101 до 200…………. 5 мест и дополнительно 3 % от количества мест свыше 100;
- от 201 до 500…………. 8 мест и дополнительно 2 % от количества мест свыше 200;
- 501 и более………. 14 мест и дополнительно 1 % от количества мест свыше 500.


Каждое выделяемое машино-место должно обозначаться дорожной разметкой и, кроме того, на участке около здания – дорожными знаками, внутри зданий – знаком доступности, 
выполняемым на вертикальной поверхности (стене, стойке и т.п.) на высоте от 1,5 до 2,0 м.
'''
```

```python

# Filters level 
Construction = {"здание" : "здание организации сферы услуг"}
Spatial_1 = {"помещение, зона" : "стоянка (парковка) транспортных средств личного пользования"}
condition_place_1 = {f"расположение {Spatial_1}" : "внутри здание организации сферы услуг"}
condition_place_2 = {f"расположение {Spatial_1}" : "около здание организации сферы услуг"}


# Definitions level
Object_1 = {'name' : 'машино-место'} #x
Object_2 = {'name' : 'машино-место для людей с инвалидностью'} #y1
Object_3 = {'name' : 'спец машино-место для людей с инвалидностью'} #y2
set_Object_1 = {f'количество {Object_1["name"]}' : 'количество машино-место'} #|X|
set_Object_2 = {f'количество {Object_2["name"]}' : 'количество машино-место для людей с инвалидностью'} #|Y1|
set_Object_3 = {f'количество {Object_3["name"]}' : 'количество спец машино-место для людей с инвалидностью'} #|Y2|

# Conditions level
condition_1 = {"B1" : "определяется как фиксированное число обязательных мест в зависимости от общего количества машиномест"}
condition_2 = {"k1" : "определяется как процентное отношение от общего количества машиномест"}
condition_3 = {"B2" : "определяется как фиксированное число обязательных мест в зависимости от общего количества машиномест"}
condition_4 = {"k2" : "определяется как процентное отношение от общего количества машиномест"}
condition_5 = {"k2" : "определяется как процентное отношение от общего количества машиномест мест свыше 100"}
condition_6 = {"k2" : "определяется как процентное отношение от общего количества машиномест мест свыше 200"}
condition_7 = {"k2" : "определяется как процентное отношение от общего количества машиномест мест свыше 500"}


# Values level

rule_1 = {"B1" : "не менее одного места", "value": 1}
rule_2 = {"k1" : "не менее 10% от общего количества машиномест", "value": 0.1}

rule_3 = {"B2" : "не менее одного места"}
rule_4 = {"k2" : "5% от общего количества машиномест"}

rule_5 = {"B2" : "5 мест"}
rule_6 = {"k2" : "3% от количества мест свыше 100"}

rule_7 = {"B2" : "8 мест"}
rule_8 = {"k2" : "2% от количества мест свыше 200"}

rule_9 = {"B2" : "14 мест"}
rule_10 = {"k2" : "1% от количества мест свыше 500"}


# Rule level

# |Y1| = max(1, |X| * 0.1)

# |Y2| = max(1, |X| * 0.05)
# |Y2| = 5 + 0.03 * (|X|-100)
# |Y2| = 8 + 0.02 * (|X|-200)
# |Y2| = 14 + 0.01 * (|X|-200)
```

```python
set_Object_2 = {f'количество {Object_2["name"]}' : 'количество машино-место для людей с инвалидностью'} #|Y1|


rule_01 = {"object_for_check" : set_Object_2, 
        "conditions" : {
            set_Object_1 : {
                "min" : min(int),
                "max" : max(int),
        }},
        "regulation" : "определяется минимальное число обязательных мест", 
        "value_text" : "не менее одного места",
        "value" : 1,
        "units" : int,
        }

rule_02 = {"object_for_check" : set_Object_2, 
         "conditions" : {
            set_Object_1 : {
                "min" : min(int),
                "max" : max(int),
        }},          
        "regulation" : "определяется как процентное отношение от общего количества машиномест", 
        "value_text" : "не менее 10% от общего количества машиномест",
        "value" : 0.1,
         "units" : float,
         }
```

```python
class Building():
    def __init__(self, construction):
        self.construction = construction


class Object_for_check():
    def __init__(self, name, place, number_parking_spaces, number_parking_spaces_disable, number_parking_spaces_disable_with_wheelchair) :
        self.name = name
        self.place = place
        self.number_parking_spaces = number_parking_spaces
        self.number_parking_spaces_disable = number_parking_spaces_disable
        self.number_parking_spaces_disable_with_wheelchair = number_parking_spaces_disable_with_wheelchair
```

```python
building_0 = Building(
    construction = "здание организации сферы услуг"
                      
                      )


object_0 = Object_for_check(
    name = "стоянка (парковка) транспортных средств личного пользования",
    place = "внутри здание организации сферы услуг",
    number_parking_spaces = 150,
    number_parking_spaces_disable = 15,
    number_parking_spaces_disable_with_wheelchair = 10
                      
                      )
```

```python

def check_disabled_parking_spaces(building, object_for_check):
    if (building.construction == "здание организации сферы услуг" and
        object_for_check.name == "стоянка (парковка) транспортных средств личного пользования" and
        (object_for_check.place == "внутри здание организации сферы услуг" or
         object_for_check.place == "около здание организации сферы услуг")
        ):

            if object_for_check.number_parking_spaces_disable < 1:
                print(f"проверка 1 - Ошибка!" +
                        f"количество машиномест для людей с инвалидностью {condition_1["B1"]}"+
                        f"количество машиномест для людей с инвалидностью {rule_1["B1"]}"
                        )
            else:
                print(f"проверка 1 - Успешно для количества машиномест для людей с инвалидностью прошла успешно")

            if object_for_check.number_parking_spaces_disable < object_for_check.number_parking_spaces * 0.1:
                print("проверка 2 - Ошибка!" +
                            f"количество машиномест для людей с инвалидностью {condition_2["k1"]}" +
                            f"{rule_2["k1"]}"
                )
            else:
                print(f"проверка 2 - Успешно для количества машиномест для людей с инвалидностью прошла успешно")

            
            if object_for_check.number_parking_spaces <= 100:
                if object_for_check.number_parking_spaces_disable_with_wheelchair < 1:
                    print(f"проверка 3 - Ошибка!" +
                          f"количество спец машиномест для людей с инвалидностью {condition_3["B2"]}" +
                            f"{rule_3["B2"]}"
                          )
                else:
                    print(f"проверка 3 - Успешно для количества спец машиномест для людей с инвалидностью прошли успешно")

                if object_for_check.number_parking_spaces_disable_with_wheelchair < object_for_check.number_parking_spaces * 0.05:
                    print("проверка 4 - Ошибка!" +
                              f"количество спец машиномест для людей с инвалидностью {condition_4["k2"]} - "+
                                f"{rule_4["k2"]}"
                              )
                else:
                    print(f"проверка 4 - Успешно для количества спец машиномест для людей с инвалидностью прошли успешно")


            if object_for_check.number_parking_spaces > 100 and object_for_check.number_parking_spaces <= 200:

                if object_for_check.number_parking_spaces_disable_with_wheelchair < ((object_for_check.number_parking_spaces - 100) * 0.03) + 5:
                    print("проверка 5 - Ошибка!" +
                              f"количество спец машиномест для людей с инвалидностью \n" +
                              f"\t{condition_3["B2"]}: " + f"{rule_5["B2"]} + \n" +
                              f"\t{condition_4["k2"]} - " + f"{rule_6["k2"]} "
                              )
                else:
                    print(f"проверка 5 - Успешно для количества спец машиномест для людей с инвалидностью прошли успешно")
    

#- от 101 до 200…………. 5 мест и дополнительно 3 % от количества мест свыше 100;
            
```

```python
check_disabled_parking_spaces(building_0, object_0)
```
